# Data Manipulation with Pandas

### Creating a DataFrame

```python
import pandas as pd
```

```python
# Create a DF object
students = pd.DataFrame()
```

```python
# assigning Values to rows and columns of the Data Frame
students['Name'] = ['Tanaya','Mansi','Jenil','Nirav']
students['Age'] = [24, 26, 23, 20]
students['Student'] = [True, False, True, True]

students
```

### Adding Data Using pd.concat()

```python
#  adding a new row means creating a new series object and appending it to the dataframe

new_student = {'Name':'Mahek','Age': 22, 'Student' : True}
```

```python
students = pd.concat([students, pd.DataFrame([new_student])], ignore_index = True)
```

```python
students
```

### Getting the Dimensions and Basic Info of the DataFrame

```python
print("Dimensions of the DataFrame :", students.shape)
print("----------------------")
print("Basic Info :", students.info())
print("----------------------")
print("Correlation :")
print(students.corr(numeric_only = True))
```

### Dropping Columns

```python
# always use axis = 1 for columns
students = students.drop('Student', axis = 1)
```

### Dropping Rows

```python
# always use axis = 0 for rows
students = students.drop(3, axis = 0)
```

```python
students
```

### Groupby()

#### 1. Grouping by one column

```python
nba = pd.read_csv(r"C:\Users\jyoti\Documents\DA Course FB\Excel\Files\nba.csv")
```

```python
nba.head()
```

```python
# lets group the data by team and calculate the total points scored by each team
team = nba.groupby('Team')
print(team.first()) # this prints the first entries in each group formed

```

```python
team.head()
```

#### 2. Grouping by Mutiple columns

```python
grouping = nba.groupby(['Team','Position'])
print(grouping.first())
```

#### 3. Using Aggregate functions within Groupby()

```python
aggregated_data = nba.groupby(['Team','Position']).agg(
        total_salary = ('Salary','sum'),
        average_salary = ('Salary','sum'),
        player_count = ('Name','count')
)

print(aggregated_data)
```

#### 4. Using Transformations Methods()

*Purpose : To apply group-specific functions while keeping the shape of the original dataset intact*

```python
# Lets rank each player within the team based on their Salaries

nba['Rank within Team'] = nba.groupby('Team')['Salary'].transform(lambda x:
                                                                 x.rank(ascending = False))

nba.head()
```

#### 5. Using .filter()

```python
# lets groupdby team and filter out players with average salary above a certain threshold
filtered_df = nba.groupby('Team').filter(lambda x: x['Salary'].mean() >= 1000000 )

filtered_df
```

### Joins

```python
#lets make two DataFrames to practice joins on
names = pd.DataFrame()
```

```python
names['id'] = [1,2,3]
names['First Name'] = ['Tanaya','Jyoti','Mishka']
names['Last Name'] = ['Dutt','Sharma','Shah']

names
```

```python
demographics = pd.DataFrame()
```

```python
demographics['Age'] = [24,59,4]
demographics['Weight'] = [51,69,14]
demographics['id'] = [1,2,3]
demographics
```

#### 1. Inner Join

```python
inner = pd.merge(names, demographics, on = 'id', how = 'inner')
```

```python
inner
```

#### 2. Left Join

```python
left = pd.merge(names, demographics, on = 'id', how = 'left')
```
