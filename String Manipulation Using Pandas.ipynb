# String Manipulation Using Pandas

```python
import pandas as pd
import numpy as np
data = { 'Name': ['Lukas', 'Sofia', 'Hiroshi', 'Marta', 'Yannis', np.nan, 'Elena'],
         'City': ['Berlin', 'Madrid', 'Tokyo', 'Warsaw', 'Athens', 'Oslo', 'Lisbon'] }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dtypes
```

```python
df_new = df.astype("string")
```

```python
df_new
```

```python
df_new.dtypes
```

### .lower() :  *converts strings to lower case*

```python
df_new['Name'].str.lower()
```

### .upper() : *converts strings to upper case*

```python
df_new['City'].str.upper()
```

### .strip() : *to remove trailing spaces*

```python
df_new['Name'].str.strip()
```

### .split() : *splits a string based on the character specified*

```python
print(df_new['Name'].str.split("a"))
```

### .len() : *to find the length of a string*

```python
print(df_new['City'].str.len())
```

### .cat() : *concatenate all strings into one*

```python
print(df_new['Name'].str.cat(sep = ","))
```

### .get_dummies() : *one-hot coding for Machine learning*

```python
print(df_new['City'].str.get_dummies())
```

```python
print(df_new['Name'].str.get_dummies())
```

### .startswith() : *finds the string that starts with a particular letter/multiple letters*

```python
print(df_new['Name'].str.startswith("L","E"))
```

### .endswith() : *finds the string that ends with a particular letter/letters*

```python
print(df_new['City'].str.endswith("s"))
```

### .replace(old,new): *replaces a string with a new string that you specify*

```python
df_new['Name'].str.replace('Lukas','El')
```

### .repeat() :*repeats the string the number of times specified*

```python
df_new['City'].str.repeat(3)
```

### .count(): *counts the number of times the specified character appears in the string*

```python
df_new['Name'].str.count('e')
```

### .findall() : *this returns the number of occurences of a specified argument*

```python
print(df_new['City'].str.findall('a'))
```

### .islower() :*checks if the string is lowercase, returns boolean value*

```python
df_new['Name'].str.islower()
```

### .isupper() :*checks if the string is uppercase, returns boolean value*

```python
df_new['City'].str.isupper()
```

### .isnumeric() :*checks if the string is numeric, returns a boolean value*

```python
df_new['Age'] = ["21","23","12","25","26","27","25"]
```

```python
df_new
```

```python
df_new['Age'].str.isnumeric()
```

### .swapcase() : *swaps lowercase to uppercase and vice-versa*

```python
print(df_new['Name'].str.swapcase())
```
