#### Task 1: Python List Calculation

```python
a = [10,20,30,40,50]

for i in range(len(a)):
    a[i] += 5

print(a)
```

#### Task 5: Element-wise Multiplication

```python
# list

x = [10,20,30]

for num in range(len(x)):
    x[num] *= 2

x
```

```python
# numpy

a = np.array([[1,2],[3,4]])
a = a * 2
a
```
