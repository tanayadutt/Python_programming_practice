#### Task 9: Category-Wise Sales

```python
Category_sum = sum(item['amount'] for item in sales if item['category'] == 'Electronics')
Category_sum2 = sum(item['amount'] for item in sales if item['category'] == 'Clothing')
print(Category_sum)
print(Category_sum2)
```

```python
# Alternative Method
category_totals = {}

for item in sales:
    category = item['category']
    amount = item['amount']

    if category in category_totals:
        category_totals[category] += amount
    else:
        category_totals[category] = amount

print(category_totals)
```

 #### Task 16: Pivot-Like Summary

```python
summary = {}

for order in sales:
    category = order['category']
    amount = order['amount']
    
    if category not in summary:
        summary[category] = {
            'order_count' :0,
            'total' : 0
        }
        
    summary[category]['order_count'] += 1
    summary[category]['total'] += amount
    

    
```

```python
for category in summary:
    summary[category]['average'] = (
    summary[category]['total']//summary[category]['order_count']
    )
```

####  Task 20: KPI Generator

```python
city_total = {}
categorys_total = {}
top_cust = ''
max_amt = 0

for order in sales:
    city = order['city']
    category = order['category']
    customer = order['customer']
    amount = order['amount']
    
    city_total[city] = city_total.get(city, 0) + amount
    
    categorys_total[category] = categorys_total.get(category, 0) + amount

    if amount > max_amt:
        max_amt = amount
        top_cust = customer
        
kpi = {
    'top_city': max(city_total, key=city_total.get),
    'top_category': max(categorys_total, key=categorys_total.get),
    'top_customer': top_cust
}

kpi      
```
