### Initializing Different Types of Array

```python
# initializing an all 0s matrix
np.zeros((2,3,4)) #(dimenions,rows,columns)
```

```python
# initializing an all 1 matrix
np.ones((3,4,1))
```

```python
# random decimal values
np.random.rand(5,2)
```

```python
# random integer values
np.random.randint(0,10, size = (2,3))
#(starting number, ending number is exclusive, and then specify the shape of the array that you want to create)
```

```python
# identity matrix is a square matrix with n rows and n columns, n is the parameter that you pass
np.identity(4)
```

```python
#repeating an array
arr = np.array([[1,2,3,4]])

rep = np.repeat(arr,2, axis = 0)

print(rep)
```

```python
#creating a specific array
output = np.ones((5,5))
print(output)

zero = np.zeros((3,3))
print(zero)

zero[1,1] = 9
print(zero)

output[1:4,1:4] = zero
print(output)
```

##### Linear Algebra

```python
# matrix multiplication just like in linear algebra
a = np.ones((3,4))
print(a)

b = np.full((4,3),2)
print(b)

# the row number of the first array should be the same as the column number for the second array for the matrix multiplication to work
np.matmul(a,b)
```

#### Statistics

```python
stats = np.array([[2,3,4],[5,6,7]])
stats
```

```python
# minimum value of an array
np.min(stats)
```

```python
# maximum value of an array
np.max(stats)
```

```python
# sum of the entire array and you can also specify an axis
np.sum(stats, axis = 0)
```

#### Reorganzing Arrays

```python
# reshaping an array
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

# reshaping changed its number of rows and columns to what you specified and what is feasible

after = before.reshape((8,1))
print(after)
```

```python
# vertically stacking array
v1 = np.array([1,2,3,4])
v2 = np.array([3,4,5,6])

np.vstack([v1,v2,v1])
```

```python
#Horizontally stacking arrays

h1 = np.array([2,3,4,5])
h2 = np.array([2,4,6,8])

np.hstack([h1,h2,h2])
```
