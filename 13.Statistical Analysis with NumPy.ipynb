####  Task 18: Loop-based Mean Calculation

```python
numbers = [10, 20, 30, 40, 50]

total = 0
count = 0

for num in numbers:
    total += num
    count += 1

mean = total / count

print(f"Mean calculated using loop: {mean}")
```

#### Task 21: Sales Analytics Summary

```python
sales_data = np.array([5000, 8500, 12000, 3000, 7000])
```

```python
print("Total:", np.sum(sales_data))
print("Mean:", np.mean(sales_data))
print("Median:", np.median(sales_data))
print("Standard Deviation:", np.std(sales_data))

greater = sales_data[sales_data > np.mean(sales_data)]
print("Above Average Sales:", greater)
```
