## Lists, Tuples and Sets

#### Task 1: Create and Inspect Sales Data

```python
daily_sales = [2000, 3000, 2300, 4500, 5400]

print(len(daily_sales))
print(daily_sales[0])
print(daily_sales[4])
daily_sales[2] = 2200
print(daily_sales)
```

#### Task 2: Clean Whitespace from Text Data

```python
cities = [" Ahmedabad ", "Mumbai ", " Delhi", "Pune "]
cleaned = [item.strip() for item in cities]
print(cleaned)
```

#### Task 3: Filter Invalid Entries

```python
values = [41, 23, 17, 88, 73, None]

valid_values = [item for item in values if item is not None]

average = sum(valid_values)/len(valid_values)

print(average)
```

#### Task 4: Identify Duplicate Records

```python
cust_ids = [1234, 2564, 2564, 3425, 3267, 6900, 1234]

len(cust_ids)

#count number of entries in the list
```

```python
unique_cust_ids = list(set(cust_ids))
unique_cust_ids

#Removed duplicate Values
```

#### Task 5: Index-Based Data Access

```python
monthly_revenue = [10000, 21000, 12000, 23000, 14257, 35729, 46901, 35188, 35628, 79570, 34255, 12000 ]

q1_revenue = monthly_revenue[0:3]
last_two_revenue = monthly_revenue[11:]

print(q1_revenue)
print(last_two_revenue)
```

#### Task 6: Convert List to Tuple for Data Safety

```python
employee = [1001, "Tanaya", "DA", 75000 ]

employee_tuple = tuple(employee)
print(employee_tuple)
```

```python
employee_tuple[2] = "HR"
print(employee_tuple)

# tuples are immutuable
```

#### Task 7: Tuple Unpacking in Analytics

```python
product = ("Ipad", 95000, 2)

product_name, price , quant = product

print(product_name)
print(price)
print(quant)
```

```python
total_sales = price * quant
print(total_sales)
```

#### Task 9: Remove Duplicate Leads

```python
email_ids = [
    "user1@gmail.com",
    "user2@gmail.com",
    "user1@gmail.com",
    "user1@gmail.com",
    "user2@gmail.com"
]

unique_emails = set(email_ids)

unique_email_list = list(unique_emails)
```

```python
print(unique_email_list)
```

#### Task 10: Membership Testing Optimization

#### Task 11: Build Category Validator

```python
statuses = {"pending", "shipped", "delivered", "cancelled"}

user_status = input("Enter order status: ")


if user_status in statuses:
    print("Valid order status")
else:
    print("Invalid order status")
```
