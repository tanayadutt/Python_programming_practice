### Iteration of Rows and Columns


```python
dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]}
 
df_iter = pd.DataFrame(dict)

df_iter
```

```python
# iterrows() will return (index,row) which is why we will be using both i and j in our for loop

for i,j in df_iter.iterrows():
    print(i,j)
    print()
```

```python
# iterrating Columns

for label,values in df_iter.items():
    print(f"Column:",label)
    print(values)
    print()
```

### Series.apply()

```python
a = pd.Series([2,4,6,8])
b = a.apply(lambda x: x / 2, convert_dtype = True)

print(b)
```

```python
#classify each mark in the Series as Pass or Fail based on a condition.

marks = pd.Series([98,76,54,100,43])

def f(x):
    return "pass" if x > 75 else "fail"

result = marks.apply(f)

print(result)

```

```python
#  lambda function to increase each salary value in the Series by 10%.

salary = pd.Series([10000, 12000, 50000])

hike = salary.apply(lambda x: x * 1.10)

print(hike)
```

# Data Cleaning

## Handling Missing Values


```python
# loading another Dataset to practice cleaning
employees = pd.read_csv(r"C:\Users\jyoti\Documents\DA Course FB\Excel\Files\employees.csv")

employees = pd.DataFrame(employees)
```

```python
employees.isnull().sum()
```

```python
# Filtering out non-missing values

nmg = pd.notnull(employees["Gender"])

nmgd = employees[nmg]

nmgd
```

```python
employees.shape
```

## Filling NaN Values

```python
# Filling the NaN values with "No gender"

employees['Gender'].fillna("No Gender", inplace = True)

employees["Gender"].isnull().sum()
```

```python
employees[10:30]
```

```python
employees["Team"] = employees["Team"].replace(to_replace = np.nan, value ="Team Unspecified")
```

```python
employees
```

###  Dropping NaN Values

#### Dropping rows with atleat one null value

```python
employees.dropna()
```

#### Dropping Rows with ALL null values

```python
employees.dropna(how = "all")
```

#### Dropping Columns with atleast one null value

```python
employees.dropna(axis = 1)
```

## Dropping Duplicates

```python
employees.drop_duplicates()
```

#### Dropping duplicated based on specific columns


```python
employees.drop_duplicates(subset = "First Name")
```

#### Keeping the last occurrence of the duplicates

```python
employees.drop_duplicates(keep = 'last')
```

## Changing Data Types if Required


#### Using .astype()

```python
employees.dtypes
```

```python
employees["Salary"] = employees["Salary"].astype(float)
```

## Dropping Columns with Empty Values

```python
df = pd.DataFrame({'FirstName': ['Vipul', 'Ashish', 'Milan'],
                            "Gender": ["", "", ""],
                            "Age": [0, 0, 0]})

df['Department'] = np.nan
print(df)

nan_value = float("NaN")
df.replace("", nan_value, inplace=True)

df.dropna(how='all', axis=1, inplace=True)

df
```

## String Manipulation using Pandas

```python


data = { 'Name': ['Lukas', 'Sofia', 'Hiroshi', 'Marta', 'Yannis', np.nan, 'Elena'],
         'City': ['Berlin', 'Madrid', 'Tokyo', 'Warsaw', 'Athens', 'Oslo', 'Lisbon'] }

df_new = pd.DataFrame(data)
print(df_new)
```

```python
df_new.dtypes
# Since they are all objects, lets change their dtype to string
```

```python
df_new = df_new.astype("string")
```

```python
df_new.dtypes
```

```python
print(df_new['Name'].str.lower())
```

```python
print(df_new['Name'].str.upper())
```

```python
# remove unwanted traling Spaces 
print(df_new['Name'].str.strip())
```

```python
# Splits string into a list of parts seperated by the separator specified
print(df_new['Name'].str.split('a'))
```

```python
# Length of the String
print(df_new['City'].str.len())
```

```python
# concatenates all strings using the specifies operator
print(df_new['City'].str.cat(sep = ','))
```

```python
# one-hot encoding for Machine Learning
print(df_new['City'].str.get_dummies())
```

```python
print(df_new["Name"].str.startswith('L'))
```

```python
print(df_new["Name"].str.endswith('a'))
```

```python
# used to replace a string with another 
print(df_new['Name'].str.replace('Elena','Emily'))
```

```python
# this method duplicates Each string a given number of times
print(df_new['City'].str.repeat(4))
```

```python
# this method counts how many times a specifies substring or pattern appears in each string
print(df_new['Name'].str.count('a'))
```

```python
# this returns a list of all the occurances/patterns given
print(df_new['Name'].str.findall('a'))
```

```python
# checks if string is lowercase
print(df_new['City'].str.islower())
```

```python
print(df_new['City'].str.isupper())
```

```python
print(df_new['City'].str.isnumeric())
```

```python
# this method swaps uppercase withlowercase string values
print(df_new['City'].str.swapcase())
```
