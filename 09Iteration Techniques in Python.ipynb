## 09Iteration Techniques in Python

#### Task 1: Simple for Loop

```python
a = [1,2,3,4,5]

for i in a:
    print(i)
```

#### Task 2: Using range()

```python
for i in range(2,11,2):
    print(i)
```

#### Task 3: Loop Over List

```python
courses = ['Python','SQL','Power BI','Excel']

for course in courses:
    print(course)
```

#### Task 4: Sum of Numbers

```python
total = 0
for num in range(1, 11):
    total += num
    print(total)
```

#### Task 5 : While loop basics

```python
num = 1

while num <= 5:
    print(num)
    num += 1
```

#### Task 6: Loop with condition

```python
nums = [1,3,5,6,7,8,9,10]

for num in nums:
    if num > 5:
        print(num)
```

#### Task 7: Using enumerate()

```python
courses = ['Python','SQL','Power BI','Excel']

for index, value in enumerate(courses):
    print(index, value)
```

#### Task 8: break Statement

```python
number = 1
while True:
    if number == 5:
        print("Loop Stopped")
        break
    print(number)
    number += 1
```

#### Task 9: continue Statement


```python
num = 1

while num <= 10:
    if num == 5:
        num += 1
        continue
    print(num)
    num += 1
```

#### Task 10 : Counting Logic

```python
value = [1000,4000,7000,9000,11000]

count = 0

for val in value:
    if val >= 5000:
        count += 1

print(count)
```

#### Task 11: Basic List Comprehension

```python
squares = []

for i in range(1,6):
    squares.append(i * i)
    
print(squares)
```

#### Task 12: Conditional List Comprehension

```python
even_num = []

for i in range (1,11):
    if i % 2 == 0:
        even_num.append(i)
    
print(even_num)
```

#### Task 13: String Transformation

```python
print([course.upper() for course in courses])
```

####  Task 14: Data Filtering

```python
amounts = [10000, 38270, 2000, 1820, 46370]

new_amounts = []

for amount in amounts:
    if amount >= 7000:
        new_amounts.append(amount)
    
print(new_amounts)
        
```

#### Task 15: Nested List Comprehension

```python
nested_list = [[1, 2, 3], [4, 5, 6]]

new_list = [num for sublist in nested_list for num in sublist]

print(new_list)
```

#### Task 16: Row-Based Data Processing

```python
orders = [
    {'order_id': 1, 'city': 'Ahmedabad', 'amount': 5000},
    {'order_id': 2, 'city': 'Delhi', 'amount': 8500},
    {'order_id': 3, 'city': 'Mumbai', 'amount': 12000}
]

for order in orders:
    print("Order", order['order_id'], "|", order['city'], "|", order['amount'])
```

#### Task 17: Running Total Calculation

```python
orders = [4500, 9500, 11100]

total = 0

for i in range(len(orders)):
    total += orders[i]
    print("After Order", i + 1, ":", total)
```

#### Task 18: Search Operation

```python
orders = [7000, 8500, 10000]

for amount in orders:
    if amount > 8000:
        print("Order Found:", amount)
        break
```

#### Task 19: Validation Loop

```python
amounts = [4500, 7600, -2000, 72000]

for amount in amounts:
    if amount < 0:
        print("Skipped Invalid Entry")
        continue
    print("Valid Amount:", amt)
```

 #### Task 20: Performance Comparison

```python
amounts = [5000, 8300, 11000,3400, 7600, 9000]

filtered= [amt for amt in amounts if amt > 8000]

print(filtered)
```

#### Task 21: Mini Analytics Processor

```python
amounts = [7000, 8000, -2000, 25000]

total = 0
count = 0
highest = 0

for amount in amounts:
    if amount < 0:       
        continue

    total += amount
    count += 1

    if amount > highest:
        highest = amount

average = total / count

result = {
    'total': total,
    'average': average,
    'highest': highest
}

print(result)
```
