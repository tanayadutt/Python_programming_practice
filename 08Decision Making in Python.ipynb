#### Task 13: Multiple Discount Rules

```python
# #Task:
# Apply discount:

# Electronics + amount ≥ 10000 → 25%
# Clothing + amount ≥ 5000 → 15%
# Otherwise → 5%

product_category = "Clothing"
amount = 5500

if product_category == "Electronics":
    amount = amount - amount * 0.25
    print(amount)
elif product_category == "Clothing":
    amount = amount - amount * 0.15 
    print(amount)
else:
    amount = amount - amount * 0.05
    print(amount)
```

#### Task 14: Performance Bonus System

```python
# Bonus rules:

# Experience ≥ 5 years AND rating ≥ 4 → 15000
# Experience ≥ 3 years → 8000
# Else → 3000

salary = 25000
experience = 4

if experience >= 5:
    salary = salary + 15000
elif experience >= 3:
    salary = salary + 8000
else:
    salary = salary + 3000
    
print(salary)
    
```

#### Task 19: Employee Payroll Decision Engine

```python
basic_salary = 60000
deduction = 5000

if deduction > 0:
    gross_salary = basic_salary + deduction
    net_salary = gross_salary - deduction
else:
    gross_salary = basic_salary
    net_salary = basic_salary

salary_details = {
    'gross_salary': gross_salary,
    'deduction': deduction,
    'net_salary': net_salary
}

print(salary_details)

```

#### Task 20: Real-World Order Processing Logic

```python
amount = 8500
is_premium_customer = True
is_urgent = True


if amount >= 5000:
    discount = "10%"
else:
    discount = "0%"

if is_premium_customer:
    delivery = "Free"
else:
    delivery = "Paid"

if is_urgent:
    priority = "High"
else:
    priority = "Normal"

result = {
    'discount': discount,
    'delivery': delivery,
    'priority': priority
}

print(result)
```
