 #### Task 1: Salary Distribution Analysis

```python
import numpy as np
```

```python
salaries = np.array([18000, 22000, 25000, 27000, 30000, 32000, 150000])

salaries
```

```python
print("Mean:", np.mean(salaries))
print("Median:", np.median(salaries))
print("Standard Deviation:", np.std(salaries))

print("Reason: Because of that one salary of 150000 being an outlier, thee median represent the salary values better thant he mean,\nbecasuee the mean is skewed")

q1 = np.percentile(salaries,25)

q3 = np.percentile(salaries,75)

IQR = q3 - q1 
print("IQR:",IQR)

lower_bound = q1 - (1.5 * IQR)
upper_bound = q3 + (1.5 * IQR)

print("Lower_bound:", lower_bound)
print("Upper_bound:", upper_bound)

outliers = salaries[(salaries < lower_bound) | (salaries > upper_bound)]
print("Outliers:", outliers)
```

 #### Task 2: Student Performance Stability

```python
marks = np.array([
    [78, 82, 85],
    [65, 70, 72],
    [90, 92, 95],
    [55, 60, 58]
])

```

```python
mean_marks = np.mean(marks, axis=1)
std_marks = np.std(marks, axis=1)

most_consistent = np.argmin(std_marks)
most_unstable = np.argmax(std_marks)

print("Mean per student:", mean_marks)
print("Standard deviation per student:", std_marks)
print("Most consistent student index:", most_consistent)
print("Most unstable student index:", most_unstable)
```

####  Task 3: Daily Website Traffic Review

```python
traffic = np.array([120, 130, 125, 128, 500, 135, 140])
```

```python
mean_traffic = np.mean(traffic)
median_traffic = np.median(traffic)

q1 = np.percentile(traffic, 25)
q3 = np.percentile(traffic, 75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

abnormal_days = traffic[(traffic < lower_bound) | (traffic > upper_bound)]

traffic_corrected = traffic.copy()
traffic_corrected[(traffic_corrected < lower_bound) | (traffic_corrected > upper_bound)] = median_traffic

new_mean = np.mean(traffic_corrected)
new_median = np.median(traffic_corrected)

print("Original Mean:", mean_traffic)
print("Original Median:", median_traffic)
print("Abnormal Traffic Days:", abnormal_days)
print("Corrected Traffic:", traffic_corrected)
print("New Mean:", new_mean)
print("New Median:", new_median)
```

#### Task 5: Percentile-Based Ranking

```python
scores = np.array([35, 40, 55, 60, 75, 80, 90, 95])
```

```python
q1 = np.percentile(scores, 25)
q2 = np.percentile(scores, 50)
q3 = np.percentile(scores, 75)

print("Below Average Score:",q1 )
print("Average Score:",q2 )
print("Above Average Score:",q3 )

below_average = scores[(scores <= q1)]
average = scores[(scores > q1) & (scores < q2)]
above_average = scores[(scores > q3)]

print("Below Average Students:", np.count_nonzero(below_average))
print("Average Students:", np.count_nonzero(average))
print("Above Average Students:", np.count_nonzero(above_average))
```
