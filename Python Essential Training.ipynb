### List Comprehensions with Filters

```python
myList = list(range(100))
```

```python
filterList = [item for item in myList if item % 10 ==2]
```

```python
print(filterList)
```

### List comprehensions with Functions

```python
myString = 'My Name is Tanaya Dutt. I live is Texas'
```

```python
myString
```

```python
myString.split()
```

```python
myString.split('.')
```

```python
def cleanWord(word):
    return word.replace('.',' ').upper()

[cleanWord(word) for word in myString.split()]
```

### Nested List Comprehensions

```python
[[cleanWord(word) for word in sentence.split()] for sentence in myString.split()]
```

# IF and ELSE

```python
#1,2,Fizz, 4,Buzz, Fizz, 7, 8 , Fizz, Buzz, 11, Fizz, 13, FizzBuzz
for n in range(1,101):
    if n % 15 == 0:
        print('FizzBuzz')
    else:
        if n % 3 == 0:
            print('Buzz')
        else:
            if n % 5 == 0:
                print('Fizz')
            else:
                print(n)
```

### Elif Statements

```python
for n in range(1,101):
    if n % 15 == 0:
        print('FizzBuzz')
    elif n % 3 == 0:
            print('Buzz')
    elif n % 5 == 0:
                print('Fizz')
    else:
         print(n)
```

### Single Line IF Statements

```python
# using if else in a one line statement
n = 3
print('Fizz' if n % 5 == 0 else n)
```

```python
#storing fizzbuzz as a function
fizzbuzz = 'Fizz' if n % 3 == 0 else n
fizzbuzz
```

### Pass

```python
wait_until = (datetime.now().second + 2) % 60

while datetime.now().second != wait_until:
    pass
print(f'We are at {wait_until} seconds')
```
