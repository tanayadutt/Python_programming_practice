## 12.Array Manipulation with NumPy

#### Task 1: Create Base Array

```python
import numpy as np
```

```python
a = np.array([10,20,30,40,50])
a
```

#### Task 2: Indexing (Positive Index)

```python
a[1]
```

#### Task 3: Negative Indexing

```python
a[-1]
```

#### Task 4: Basic Slicing

```python
a[1:4]
```

#### Task 5: Step Slicing

```python
result = a[::2]
print(result)
```

#### Task 6: Create 2D Array

```python
# creating a 2D array

b = np.array([[1,2,3],[4,5,6]])
b
```

#### Task 7: 2D Indexing

```python
b[1,1]
```

#### Task 8: Row & Column Slicing

```python
print(b[0])
print(b[:,1])
```

#### Task 9: Reshape Array

```python
onearray = np.array([1,2,3,4,5,6])

twoarray = onearray.reshape([3,2])
print(twoarray)
```

#### Task 10: Flatten Array

```python
arr = np.array([[1,2,3,4,5],[2,3,4,5,6]])


new_arr = arr.flatten()
print(new_arr)
```

#### Task 11: Vertical Stack 

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

np.vstack([a,b])
```

 #### Task 12: Horizontal Stack

```python
np.hstack([a,b])
```

#### Task 13: Column Stack

```python
# will likely stack columns
stacked = np.column_stack([a,b])
print(stacked)
```

 #### Task 14: Stack with stack()

```python
stack = np.stack((a,b), axis = 0)
print(stack)
```

####  Task 15: Split Array into Equal Parts

```python
# spliting array into three equal parts

arr = np.array([1,3,5,7,9,11])

new_arr = np.array_split(arr,3)

print(new_arr)
```

#### Task 17: Vertical Split

```python
c = np.array([[1,2,3],[3,4,5],[4,5,6]])

v_split = np.vsplit((c),3)
print(v_split)
```

#### Task 18: Horizontal Split

```python
c = np.array([[1,2,3],[3,4,5],[4,5,6]])

h_split = np.hsplit((c),3)
print(h_split)
```

#### Task 19: Element-wise Addition

```python
d = c + 10
print(d)
```

#### Task 20: Element-wise Multiplication

```python
e = c * d
print(e)
```

 #### Task 21: Conditional Vectorized Operation

 #### Task 22: Vectorized Filtering
